```python
ruta_archivo = '/Users/sandradiaz/Documents/CaptstoneProject/dades/2022/BicingNou_ESTACIONS/2022_03_Marc_BicingNou_ESTACIONS.csv'

dataset = pd.read_csv(ruta_archivo)
```

# Variables

```python
dataset.columns
```

```python
#formato
dataset['last_updated'] = pd.to_datetime(dataset['last_updated'])
```

```python
dataset['last_updated']
#datetime64[ns] = formato fecha y hora
```

```python
#orden ascendente
dataset.sort_values('last_updated', inplace=True)
```

```python
#disponibilidad hora anterior con shift - desplazan los valores de esa columna hacia abajo en una posición-
dataset['timestamp_1'] = dataset['last_updated'].shift(1)
```

```python
#disponibilidad dos horas antes
dataset['timestamp_2'] = dataset['timestamp_1'].shift(1)
```

```python
#disponibilidad tres horas antes
dataset['timestamp_3'] = dataset['timestamp_2'].shift(1)
```

```python
#disponibilidad cuatro horas antes
dataset['timestamp_4'] = dataset['timestamp_3'].shift(1)
```

```python
dataset.head(15)
```

```python
dataset['pct_bikes_doc'] = dataset['num_bikes_available'] / dataset['num_docks_available']
```

```python
dataset.head(2)
```

```python
#disponibilidad bicis hora anterior
dataset['pct_bikes_doc_1'] = dataset['pct_bikes_doc'].shift(1)
```

```python
#disponibilidad bicis 2 horas anteriores 
dataset['pct_bikes_doc_2'] = dataset['pct_bikes_doc_1'].shift(1)
```

```python
#disponibilidad bicis 3 horas anteriores 
dataset['pct_bikes_doc_3'] = dataset['pct_bikes_doc_2'].shift(1)
```

```python
#disponibilidad bicis 4 horas anteriores 
dataset['pct_bikes_doc_4'] = dataset['pct_bikes_doc_3'].shift(1)
```
